```python
from sklearn.decomposition import PCA
#PCA transformation on the physical activity group train dataset
pca1 = PCA()
project_phy_train = pca1.fit_transform(norm_phy_train)

#PCA transformation ont the no activity group train dataset
pca2 = PCA()
eigen_decomp = pca2.fit(norm_na_train)
project_na_train = eigen_decomp.transform(norm_na_train)
```

```python
import numpy as np
print(f'physical activity variance: {np.round(pca1.explained_variance_, 1)}')
print(f'physical activity variance percentage: {np.round(pca1.explained_variance_ratio_, 1)}')
print()
print(f'No physical activity variance: {np.round(pca2.explained_variance_, 1)}')
print(f'No physical activity variance percentage: {np.round(pca2.explained_variance_ratio_, 1)}')
```

```python
pca_table =  pd.DataFrame({
    'Principal component': [f'PC{i+1}' for i in range(len(pca1.explained_variance_ratio_))],
    'NoActivity Variance (%)': [round(v*100, 2) for v in pca2.explained_variance_ratio_],
    'PhysActivity Variance (%)': [round(v*100, 2) for v in pca1.explained_variance_ratio_]
})
print(pca_table)
```

```python
var = np.sum(pca1.explained_variance_[0:2])
var_percent = np.sum(pca1.explained_variance_ratio_[0:2])*100
print(f'physical activity variance: {var}')
print(f'physical activity percentage variance: {var_percent}%')
print()
var = np.sum(pca2.explained_variance_[0:2])
var_percent = np.sum(pca2.explained_variance_ratio_[0:2])*100
print(f'No physical activity variance: {var}')
print(f'No physical activity percentage variance: {var_percent}%')
```

## Task2

```python
from sklearn.model_selection import train_test_split
data = db_na_train.iloc[:, 1:]
target = db_na_train.Diabetes_binary
x_na_train, x_na_test, y_na_train, y_na_test = train_test_split(data, target, test_size=0.3, random_state=42)
```

```python
print(f'small trian II size: {x_na_train.shape}')
print(f'small test II size: {x_na_test.shape}')
```

```python
#normalising the no activity train dataset II
scaler = StandardScaler().fit(x_na_train)
xnorm_na_train = scaler.transform(x_na_train)

print(xnorm_na_train.mean(axis=0)) # print the mean value of each feature  in no activity train normalized set
print(xnorm_na_train.std(axis=0)) # print the standard deviation value of each feature in no activity train normalized set
print()

xnorm_na_test = scaler.transform(x_na_test)
print(xnorm_na_test.mean(axis=0)) # print the mean value of each feature  in no activity test normalized set
print(xnorm_na_test.std(axis=0)) # # print the std value of each feature  in no activity test normalized set
```

```python
#training on the  no activity train II dataset
#C=1, gamma=1
from sklearn.svm import SVC
from sklearn.metrics import classification_report
svc1 = SVC(kernel='rbf', class_weight='balanced', C=1, gamma=1)
model1 = svc1.fit(xnorm_na_train, y_na_train)

"""the fitted model should be validated on the scaled validation set"""
yfit1 = model1.predict(xnorm_na_test)

"""performance measurement"""
print(classification_report(y_na_test, yfit1))
```

```python
#training on the  no activity train II dataset
#C=5, gamma=0.5
svc2 = SVC(kernel='rbf', class_weight='balanced', C=5, gamma=0.5)
model2 = svc2.fit(xnorm_na_train, y_na_train)

"""the fitted model should be validated on the scaled validation set"""
yfit2 = model2.predict(xnorm_na_test)

"""performance measurement"""
print(classification_report(y_na_test, yfit2))
```

```python
#training on the  no activity train II dataset
#C=0.5, gamma=0.05
svc3 = SVC(kernel='rbf', class_weight='balanced', C=0.5, gamma=0.05)
model3 = svc3.fit(xnorm_na_train, y_na_train)

"""the fitted model should be validated on the scaled validation set"""
yfit3 = model3.predict(xnorm_na_test)

"""performance measurement"""
print(classification_report(y_na_test, yfit3))
```

```python
#normalising the whole no activity training and test dataset
values = db_na_train.iloc[:, 1:]
scaler = StandardScaler().fit(values)
xnorm_na_train = scaler.transform(values)
values = db_na_test.iloc[:, 1:]
xnorm_na_test = scaler.transform(values)
```

```python
#training on the whole no activity  training and test dataset
#C=0.5, gamma=0.05
svc = SVC(kernel='rbf', class_weight='balanced', C=0.5, gamma=0.05)
model_na = svc.fit(xnorm_na_train, db_na_train.Diabetes_binary)

"""the fitted model should be validated on the scaled validation set"""
yfit = model_na.predict(xnorm_na_test)

"""performance measurement"""
print(classification_report(db_na_test.Diabetes_binary, yfit))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
conf_matrix = confusion_matrix(db_na_test.Diabetes_binary, yfit)
# Plot the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=[0, 1])
disp.plot()
plt.title('Confusion matrix for no activity test dataset')
plt.show()
```
